```python
import pandas as pd
fdata = pd.read_csv('faculties_.csv')
cdata = pd.read_csv('class.csv')
```

```python
fdata
```

```python
profiles = list(cdata['биология/медицина'])
profiles.append('биология/медицина')
profiles
```

```python
assump = []
for faculty in fdata['universities_faculty_name']:
    print(faculty)
    if faculty.find('математ') > 0:
        assump.append('математика')
    elif faculty.find('хими') > 0:
        assump.append('химия')
    elif faculty.find('географ') > 0:
        assump.append('география')
    elif faculty.find('физи') > 0:
        assump.append('физика')
    elif faculty.find('биолог') > 0 or faculty.find('медицин') > 0 or faculty.find('био') > 0 or faculty.find('экологич') > 0:
        assump.append('биология/медицина')
    elif faculty.find('техн') > 0 or faculty.find('комп') > 0:
        assump.append('техника и технологии')
    elif faculty.find('менеджмент') > 0 or faculty.find('управл') > 0:
        assump.append('менеджмент/управление')
    elif faculty.find('филол') > 0:
        assump.append('филология')
    elif faculty.find('филос') > 0:
        assump.append('философия')
    elif faculty.find('искус') > 0 or faculty.find('дизайн') > 0 or faculty.find('культур') > 0:
        assump.append('искусство/культура')
    elif faculty.find('педагог') > 0 or faculty.find('пcихол') > 0 or faculty.find('препод') > 0:
        assump.append('психология/педагогика')
    elif faculty.find('прав') > 0 or faculty.find('юриспруд') > 0:
        assump.append('юриспруденция')
    elif faculty.find('иностран') > 0 or faculty.find('язык') > 0 or faculty.find('лингвист') > 0 or faculty.find('перевод') > 0 or faculty.find('международ') > 0:
        assump.append('иностранные языки')
    elif faculty.find('строител') > 0:
        assump.append('строительство')
    elif faculty.find('истор') > 0 or faculty.find('полит') > 0:
        assump.append('история/политология')
    elif faculty.find('общест') > 0 or faculty.find('социо') > 0:
        assump.append('обществознание/cоциология')
    elif faculty.find('маркетинг') > 0:
        assump.append('маркетинг/PR')
    elif faculty.find('эконом') > 0 or faculty.find('финан') > 0:
        assump.append('экономика/финансы')
    elif faculty.find('бизнес') > 0:
        assump.append('бизнес/предпринимательство')
    elif faculty.find('естеств') > 0:
        assump.append('естественные науки')
    else:
        assump.append('Not defined')
```

```python
len(assump)
```

```python
newdata = pd.DataFrame({'universities_faculty_name':fdata['universities_faculty_name'], 'profiles':assump})
```
