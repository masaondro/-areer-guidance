```python
def median(lst):
    n = len(lst)
    if n < 1:
            return None
    if n % 2 == 1:
            return sorted(lst)[n//2]
    else:
            return sum(sorted(lst)[n//2-1:n//2+1])/2.0
```

```python
def publicthreshold(freq, threshold = 30):
    lst = []
    for elem in freq.items():
        num = elem[1]
        if num >= threshold:
            lst.append(elem[0])
    return lst
```
